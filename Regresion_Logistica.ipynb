```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
#Cargar desde seaborn
df=sns.load_dataset('titanic')
df.head(20)
```

```python
#Rellenar nulos
df = df.fillna(method='bfill')
df = df.fillna(method='ffill')
```

```python
#Declaramos las variables dependientes e independientes para la regresión logística
Vars_Indep = df [['pclass', 'age', 'fare']]
#La dependiente tiene que ser dicotómica
Vars_Dep = df ['alive']
```

```python
#Redefinimos las variable
X = Vars_Indep
y = Vars_Dep
```

```python
#Dividimos los datos en entrenamiento y prueba
#
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)
```

```python
#Se escalan los datos
escalar = StandardScaler()
```

```python
#Para realizar el escalamiento de las variables "X" tanto de entrenamiento como de prueba
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
#y no se puede porque no es un variable numerica
```

```python
#Definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
#Entrenamos el algoritmo
algoritmo.fit(X_train, y_train)
```

```python
#Realizamo una predicción
y_pred = algoritmo.predict(X_test)
y_pred
```

```python
#Verificamos la matriz de confusión
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
```

```python
#Calculamos la precisión del modelo
#pos_label es la etiqueta binaria en este caso es "yes" tambien se puede poner "no"
from sklearn.metrics import precision_score

presicion = precision_score(y_test, y_pred, average="binary", pos_label="yes")
print('Precisión del modelo:')
print(presicion)
```

```python
#Calculamos la exactitud del modelo
from sklearn.metrics import accuracy_score

exactitud = accuracy_score(y_test, y_pred)
print('Exactitud del modelo:')
print(exactitud)
```

```python
#Calculamos la sensibilidad del modelo
from sklearn.metrics import recall_score

sensibilidad = recall_score(y_test, y_pred, average="binary", pos_label="no")
print('Sensibilidad del modelo:')
print(sensibilidad)
```
